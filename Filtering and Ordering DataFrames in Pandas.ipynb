#### Filtering and ordering **`DataFrame`** in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r'world_population.csv')
```

```python
df.head()
```

```python
filter_df = df[df['Rank'] <=150]
filter_df
```

```python
speciefic_countries= ['Bangladesh','Brazil','India','Afghanistan','Algeria','Vietnam	']
filter_df[filter_df['Country'].isin(speciefic_countries)]
```

```python
filter_df[filter_df['Country'].str.contains('United')]
```

```python
filter_df[filter_df['Continent'].str.contains('Europe')]
```

# Filter Rows in DataFrame
> there is two main way :<br>
1-using filter with `like` or `items` arguments<br>
2- using `loc` & `iloc`

```python
df2 = df.set_index('Country')
df2= df2.filter(like = "Population",axis = 1)
df2 = df2.filter(like = 'United',axis = 0)
df2
```

```python
df2.filter(items = ['United Arab Emirates','United States'],axis = 0)
```

```python
df2.loc['United Arab Emirates']                 # take string variables
```

```python
df2.iloc[3]         # take index
```

```python
df2.iloc[1:3,3:6]
```

```python
filter_df[filter_df['Rank'] <20].sort_values(by= 'Rank' ,ascending= True)
```
